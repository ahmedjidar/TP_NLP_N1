> **Mise en minuscules**

```python
data["text_lower"] = data["text"].str.lower()
data.head()
```

****Suppression des ponctuations****

```python
import string

# Punctuation list
punct = string.punctuation

# Function to clean text
def clean_text(text):
    return text.translate(str.maketrans('', '', punct))

# Apply function to 'text' column
data['clean_text'] = data['text'].apply(clean_text)

# Show updated dataframe
data.head()

```

**Suppression des mots vides**

```python
from nltk.corpus import stopwords

# English stopwords
stop_words = set(stopwords.words('english'))

# Function to clean text
def clean_text(text):
    return ' '.join(word for word in text.split() if word not in stop_words)

# Apply function to 'punc_remove' column
data['clean_text'] = data['punc_remove'].apply(clean_text)

# Show updated dataframe
data.head()

```

**Suppression des mots frÃ©quents**

```python
from collections import Counter

# Counter object
counter = Counter()

# Count words in 'stopwords_remove' column
for text in data['stopwords_remove']:
    counter.update(text.split())

# Top 10 words
top_words = counter.most_common(10)

print(top_words)

```

**Suppression des mots rares**

```python
from collections import Counter

# Counter object
counter = Counter()

# Count words in 'stopwords_remove' column
for text in data['stopwords_remove']:
    counter.update(text.split())

# Frequency threshold
freq_threshold = 5

# Common words
common_words = {word for word, count in counter.items() if count >= freq_threshold}

# Function to filter text
def filter_text(text):
    return ' '.join(word for word in text.split() if word in common_words)

# Apply function to 'stopwords_remove' column
data['filtered_text'] = data['stopwords_remove'].apply(filter_text)

# Show updated dataframe
data.head()
```

**Stemming**

```python
from nltk.stem.porter import PorterStemmer

# PorterStemmer object
stemmer = PorterStemmer()

# Function to stem words
def stem_text(text):
    return ' '.join(stemmer.stem(word) for word in text.split())

# Apply function to 'text' column
data['stemmed_text'] = data['text'].apply(stem_text)

# Show updated dataframe
data.head(2)

```

**Lemmatisation**

```python
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

# Lemmatizer object
lemmatizer = WordNetLemmatizer()

# Function to lemmatize words
def lemmatize_text(text):
    return ' '.join(lemmatizer.lemmatize(word) for word in text.split())


```

**Suppression des URL, Suppression des balises**

```python
import re

# Function to strip URLs
def strip_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)

# Function to strip HTML tags
def strip_html(text):
    return re.sub(r'<.*?>', '', text)

# Example:
text = "Check out my website at https://example.com! <p>This is a paragraph.</p>"
clean_text = strip_html(strip_urls(text))
print(clean_text)
```
